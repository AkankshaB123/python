# **Filtering Assignment**
1. Display Data for all Private Rooms in the Central Region that have at least 10 reviews, cost more than 50.
2. Display all the listings of the top host in the data
3. Find the distribution of room_types for the most listed neighborhood.

```python
listings[ (listings['room_type'] == 'Private room') &
          (listings['neighbourhood_group'] == 'Central Region') &
          (listings['number_of_reviews'] >= 10) &
          (listings['price'] > 50)
]
```

```python
#Display all the listings of the top host in the data
listings[listings['host_name'] == listings['host_name'].value_counts().index[0]]
```

```python
#3.Find the distribution of room_types for the most listed neighborhood
#Nested concept
listings[listings['neighbourhood'] == listings['neighbourhood'].value_counts().index[0]]['room_type'].value_counts()
```
