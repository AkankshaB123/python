# Task - 2

* Let's create 3 sets
* s1 = {1,2,3,4,5}
* s2 = {3,8,5,9,10}
* s3 = {1,2,9}
* Add 100 to each one of them
* Find their intersection: s1 and s2, s2 and s3, s3 and s1
* Is s3 a subset of (s1 union s2)?
* Find the NET sorted values using all elements in all 3 sets - Descending
* Convert the sets after union to a lit and then a tuple and access the last value

```python
s1 = {1,2,3,4,5}
s2 = {3,8,5,9,10}
s3 = {1,2,9}
```

```python
s1.add(100)
s2.add(100)
s3.add(100)
```

```python
print(s1.intersection(s2))
print(s2.intersection(s3))
print(s3.intersection(s1))
```

```python
s3.issubset(s1.union(s2))
```

```python
#Find the NET sorted values using all elements in all 3 sets - Descending
sorted(s1.union(s2).union(s3), reverse = True)
```

```python
#Convert the sets after union to a list and then a tuple and access the last value
list(tuple(s1.union(s2).union(s3)))[-1]
```

```python
list(s1.union(s2).union(s3))
```

```python
tuple(s1.union(s2).union(s3))
```
