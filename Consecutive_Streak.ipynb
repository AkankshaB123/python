```python
data = [
    ("A", "2024-01-01"),
    ("A", "2024-01-02"),
    ("A", "2024-01-03"),
    ("A", "2024-01-06"),
    ("A", "2024-01-07"),
    ("B", "2024-02-10"),
    ("B", "2024-02-11"),
]
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=["name", "date"])
df
```

# **Sort**

```python
df = df.sort_values(by=["name", "date"])
df
```

# **Find if previous date is consecutive**

```python
df['previous'] = df.groupby('name')['date'].shift(1)
df
```

```python
df['dt'] = pd.to_datetime(df['date'])
df['previous'] = pd.to_datetime(df['previous'])
df
```

```python
df['new_streak'] = (df['dt'] - df['previous']).dt.days!=1
df
```

# **Cumulative Count -> Streak Group**

```python
df['streak_grp'] = df.groupby('name')['new_streak'].cumsum()
df
```

# **Aggregate**

```python
result = df.groupby(['name', 'streak_grp']).agg(streak_start=('date', 'min'), streak_end=('date', 'max'), streak_length=('streak_grp', 'count')).reset_index()
result
```
