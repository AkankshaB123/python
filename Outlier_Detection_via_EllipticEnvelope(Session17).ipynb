```python
#Load libraries
import numpy as np
from sklearn.covariance import EllipticEnvelope
from sklearn.datasets import make_blobs #Create artifical data based on the requirement
import matplotlib.pyplot as plt
```

```python
#Create simulated data
features, _ = make_blobs(n_samples = 10, #rows
                         n_features = 2, #columns
                         centers = 1, #center of data
                         random_state = 1) #random value to initialise
features
```

```python
plt.scatter(features[:,0],features[:,1])
#Splot circle being built with 1000 samples
```

```python
#Replace the first observations with 100
features[0,0] = 100
features[0,1] = 100
features
#See value differences - outlier addition
```

```python
#Create detector
outlier_detector = EllipticEnvelope(contamination = .1) #10% of data is outliers
#Contamination is the proportion of observations that are outliers, more contamination larger ellipse it becomes
```

```python
#Fit detector
outlier_detector.fit(features)
```

```python
#Predict outliers
outlier_detector.predict(features)
#outlier : -1
#Next step: transform it into a column and filter outliers
```
