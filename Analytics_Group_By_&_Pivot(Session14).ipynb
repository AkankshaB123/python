#For Entry-level Data Science: Pivot Table and Group are very necessary

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/listings.csv')
df.head(3)
```

# **Group By Task**
1. Mean Price of every room type in a neighbourhood
2. What is the count of room types by host_name
3. Value counts of Neighnourhood Group using groupby
4. Count of hotels in every neighbourhood group and neighbourhood
5. Most Expensive Neighbourhood Group
6. Cheapest Neighbourhood
7. Neighbourhood with the maximum_minimum_nights
8. Least Popular Neighbourhood_group

```python
#1. Mean Price of every room type in a neighbourhood
df.groupby(['neighbourhood','room_type']).agg({'price':'mean','number_of_reviews':'mean'})
```

```python
#2. What is the count of room types by host_name
df.groupby('host_name').agg({'room_type':'count'}).head(3)
```

```python
#3. Value counts of Neighnourhood Group using groupby
df.groupby('neighbourhood_group').agg({'neighbourhood':'count'})
```

```python
#4.Count of hotels in every neighbourhood group and neighbourhood
df.groupby(['neighbourhood_group','neighbourhood']).agg({'id':'count'})
```

```python
#5.Most Expensive Neighbourhood Group
df.groupby('neighbourhood_group').agg({'price':'mean'}).sort_values('price',ascending=False)
```

```python
df.groupby('neighbourhood_group').agg({'price':'mean'}).idxmax()
```

```python
#6.Cheapest Neighbourhood
df.groupby('neighbourhood_group').agg({'price':'mean'}).sort_values('price',ascending=True).head(1)
```

```python
#7.Neighbourhood with the maximum_minimum_nights
df.groupby('neighbourhood').agg({'minimum_nights':'mean'}).sort_values('minimum_nights',ascending=False).head(3)
```

```python
#8.Least Popular Neighbourhood_group
df.groupby('neighbourhood_group').agg({'number_of_reviews':'sum'}).sort_values('number_of_reviews',ascending=True).head(3)
```

# **Pivot Table**

```python
df.groupby(['neighbourhood_group','room_type']).agg({"price":"mean"})
```

```python
#Use-case: To change above view to this view.
df.pivot_table(index='neighbourhood_group',
               columns='room_type',
               values='price',
               aggfunc='mean')
#4 parts of a pivot_table: index, columns, values, aggfunc
```

```python
df.pivot_table(index=['neighbourhood_group','neighbourhood'],
               columns='room_type',
               values=['price',"number_of_reviews"],
               aggfunc=['mean','min'])
```

# **Task**
1. For each neighbourhood_group and neighbourhood by Room Type what is the average minimum nights offered by the hotels
2. Which neighbourhood has the minimum price for a room type?
3. How many unique hosts in each neighbourhood group by room type

```python
#For each neighbourhood_group and neighbourhood by Room Type what is the average minimum nights offered by the hotels
df.pivot_table(index=['neighbourhood_group','neighbourhood'],
               columns='room_type',
               values='minimum_nights',
               aggfunc='mean')
```

```python
#2. Which neighbourhood has the minimum price for a room type?
df.pivot_table(index='neighbourhood',
               columns='room_type',
               values='price',
        aggfunc='min').sort_values(by="Entire home/apt")
        #Check for Private room, then for Shared room
```

```python
#3. How many unique hosts in each neighbourhood group by room type
df.pivot_table(index='neighbourhood_group',
               columns='room_type',
               values='host_id',
               aggfunc='nunique')
```
