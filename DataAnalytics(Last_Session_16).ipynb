```python
#Rename Columns
df.rename(columns={'SUBDIVISION': 'SUB_DIVISION'}, inplace=True)
df.columns
```

```python
df['GPU'].nunique()
```

```python
#Values with pattern 'Yangten' in GPU column
df[df['GPU'].str.contains('Yangten', na=False)]['GPU'].value_counts()
```

```python
#Replace certain values Ex: in GPU column: Yangten, Yngteen
df['GPU'] = df['GPU'].replace({'Yangten': 'Yngteen'})
df[df['GPU'].str.contains('Yngteen', na=False)]['GPU'].value_counts()
```

```python
#Find Outliers in the Data
df.describe()
```

```python
#Convert all column names to lower cases
df.columns = df.columns.str.lower()
df.head(2)
```

```python
#Find out the Proportion of Males and Females in the Total Population and each row
df['male_proportion'] = df['male'] / df['total']
df['female_proportion'] = df['female'] / df['total']
df
```

```python
#Find the most populated Subdivision
df.groupby('sub_division')['total'].sum().sort_values(ascending=False).head(1)
```

```python
#Based on District Find Sex Ratio
df.groupby('district')[['male', 'female']].mean()
```

```python
#Replace District Column values with N,E,W,S
df['district'].replace({'North': 'N', 'East': 'E', 'West': 'W', 'South': 'S'})
```
