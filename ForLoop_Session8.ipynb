```python
#Factorial
num = input("Enter a number: ")
num = int(num)
factorial = 1
for i in range(1,num + 1):
    factorial = factorial*i
    # 1*5 = 5
    # 5*4 = 20
    # 20*3 = 60
    # 60*2 = 120
    # 120*1 = 120
print("The factorial of",num,"is",factorial)
```
