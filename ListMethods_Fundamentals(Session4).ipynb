Continguous & Non-Contiguous Memory allocation **(Depending on the size of the list, python is doing this in the backend)** - To optimize for speed and security in data
* 100 MB requirement
* 2 ways:
1. Find complete empty storage/patch - straight/in-a-sequence, this is "contiguous memory allocation".
2. If it's not sequentially available, that's non-contiguous memory allocation

Note:
1. List - collection of objects
2. Every data type should provide 3 functionalities:
* Functionality to *store* values eg. l1 = [1,2,3]
* Functionality to *modify/update* values eg. existing list l1, update l1[0]
* Functionality to *retrieve/access* values eg. existing list l1, update l1[0]
3. Use Python methods efficiently to make the python EDA faster

* Append method - add elements to the list one at a time, hacky way possible but becomes a nested list
* extend method - add multiple elements
* If in case of .sort method - can never go back to the original list
Concepts to be aware about:
1. Permanent  or in place change
2. Temporary or out of place change
* Prefer using ***sorted(l1)*** - but does not change l1 permanently if permanently needed -> l1 = sorted(l1)
