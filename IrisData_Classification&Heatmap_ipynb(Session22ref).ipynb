```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
irisdata = pd.read_csv("/content/sample_data/iris.csv")
irisdata.head(5)
```

```python
for class_ in irisdata['variety'].unique():
  print(class_)
```

```python
for class_ in irisdata['variety'].unique():
  df_class = irisdata[irisdata['variety'] == class_].copy()
  df_class.loc[:, 'ratio'] = df_class['sepal.length'] / df_class['sepal.width']
  plt.scatter(df_class.index, df_class['ratio'], label = class_)

plt.legend()
plt.show()
```

Ratio is the independent variable with which variety of the flower can be identified

```python
corr = irisdata.drop('variety', axis=1).corr()
```

```python
import seaborn as sns
sns.heatmap(corr, annot = True, cmap = 'rocket')
plt.show()
```

Identify highly correlated variables and remove one from each correlated pair, as **multicollinearity** adds redundant information and unnecessarily increases computational cost without improving model performance.
