```python
import pandas as pd
df = pd.read_csv("/content/sample_data/titanic3.csv")
df.head(3)
```

***Filtering Assignment***
1. How many voyagers have paid greater than the mean fare value?
2. How many voyagers paying greater than mean fare are female?
3. How many male voyagers survived?
4. How many female voyagers didn't survive?
5. How many females survived?

Note: Do not hard-code anything. Big tech or large scale companies prefer that

```python
#1. How many voyagers have paid greater than the mean fare value?
#Remember this: within the dataframe filtering
len(df[df['fare']>df['fare'].mean()])
```

```python
#2. How many voyagers paying greater than mean fare are female?
#Remember to have this () within df [] to have multiple conditions
len(df[(df['fare']>df['fare'].mean()) & (df['sex'] == 'female')])
```

```python
#3. How many male voyagers survived?
len(df[df['sex'] == 'male'])
```

```python
len(df[df['sex'] == 'female'])
```

```python
#How many female voyagers didn't survive?
len(df[(df['sex'] == 'female') & (df['survived'] == 0.0)])
```

```python
#How many females survived?
len(df[(df['sex'] == 'female') & (df['survived'] == 1.0)])
```
